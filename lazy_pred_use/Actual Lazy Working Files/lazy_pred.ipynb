Makinng the merged dataset and mapping it to the earthquake dataset

```python
import pandas as pd

# Load datasets
file_earthquake = "acceleration_stats_1sec_samples.csv"
file_noise = "noise_acceleration_stats_6sample_samples.csv"

df_earthquake = pd.read_csv(file_earthquake)
df_noise = pd.read_csv(file_noise)

# Manually construct noise data in the proper format by stacking S1-S6 as separate rows
noise_rows = []

for _, row in df_noise.iterrows():
    for i in range(1, 7, 2):  # Iterating in pairs (S1-S2, S3-S4, S5-S6)
        new_row = {
            "Before Mean": row[f"S{i} Mean"], "Before Median": row[f"S{i} Median"], "Before Mode": row[f"S{i} Mode"],
            "Before Std Dev": row[f"S{i} Std Dev"], "Before Skewness": row[f"S{i} Skewness"], "Before Kurtosis": row[f"S{i} Kurtosis"],
            "Before Variance": row[f"S{i} Variance"], "Before Max": row[f"S{i} Max"], "Before Min": row[f"S{i} Min"], "Before Z-Score": row[f"S{i} Z-Score"],

            "After Mean": row[f"S{i+1} Mean"], "After Median": row[f"S{i+1} Median"], "After Mode": row[f"S{i+1} Mode"],
            "After Std Dev": row[f"S{i+1} Std Dev"], "After Skewness": row[f"S{i+1} Skewness"], "After Kurtosis": row[f"S{i+1} Kurtosis"],
            "After Variance": row[f"S{i+1} Variance"], "After Max": row[f"S{i+1} Max"], "After Min": row[f"S{i+1} Min"], "After Z-Score": row[f"S{i+1} Z-Score"],

            "earthquake": 0  # Label for noise data
        }
        noise_rows.append(new_row)

# Convert list to DataFrame
df_noise_fixed = pd.DataFrame(noise_rows)

# Add labels for earthquake data
df_earthquake["earthquake"] = 1

# Select only necessary columns from earthquake dataset
selected_columns = list(df_noise_fixed.columns)
df_earthquake_cleaned = df_earthquake[selected_columns]

# Combine datasets
df_combined = pd.concat([df_earthquake_cleaned, df_noise_fixed], ignore_index=True)

# Save the cleaned dataset
df_combined.to_csv("merged_dataset.csv", index=False)

print("Dataset merged successfully and saved as 'merged_dataset.csv'.")
```

```python
# Install necessary libraries if not installed
# pip install pandas numpy scikit-learn lazypredict 

import pandas as pd
from sklearn.model_selection import train_test_split
from lazypredict.Supervised import LazyClassifier

# Load both datasets
file_earthquake = "acceleration_stats_1sec_samples.csv"  # Update with actual path
file_noise = "noise_acceleration_stats_6sample_samples.csv"  # Update with actual path

df_earthquake = pd.read_csv(file_earthquake)
df_noise = pd.read_csv(file_noise)

# Add labels: 1 for earthquake, 0 for noise
df_earthquake["earthquake"] = 1
df_noise["earthquake"] = 0

# Combine datasets
df_combined = pd.concat([df_earthquake, df_noise], ignore_index=True)

# Drop non-feature columns
df_combined = df_combined.drop(columns=["File", "Trace ID", "P-Pick Time"], errors="ignore")

# Separate features and target
X = df_combined.drop(columns=["earthquake"])
y = df_combined["earthquake"]

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and run LazyClassifier
clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
models, predictions = clf.fit(X_train, X_test, y_train, y_test)

# Display results
print(models)
```
