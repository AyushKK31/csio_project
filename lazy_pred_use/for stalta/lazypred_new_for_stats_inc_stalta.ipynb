```python
import pandas as pd

# Load datasets
file_earthquake = "acceleration_stats_1sec_samples_with_sta_lta_mer.csv"
file_noise = "noise_acceleration_stats_6sample_samples_with_sta_lta_mer.csv"

df_earthquake = pd.read_csv(file_earthquake)
df_noise = pd.read_csv(file_noise)

# Manually construct noise data in the proper format by stacking S1-S6 as separate rows
noise_rows = []

for _, row in df_noise.iterrows():
    for i in range(1, 7, 2):  # Iterating in pairs (S1-S2, S3-S4, S5-S6)
        new_row = {
            "Before Mean": row[f"S{i} Mean"], "Before Median": row[f"S{i} Median"], "Before Mode": row[f"S{i} Mode"],
            "Before Std Dev": row[f"S{i} Std Dev"], "Before Skewness": row[f"S{i} Skewness"], "Before Kurtosis": row[f"S{i} Kurtosis"],
            "Before Variance": row[f"S{i} Variance"], "Before Max": row[f"S{i} Max"], "Before Min": row[f"S{i} Min"], 
            "Before Z-Score": row[f"S{i} Z-Score"], "Before Mean STA/LTA": row[f"S{i} Mean STA/LTA"],
            "Before Max STA/LTA": row[f"S{i} Max STA/LTA"], "Before Mean MER": row[f"S{i} Mean MER"], "Before Max MER": row[f"S{i} Max MER"],

            "After Mean": row[f"S{i+1} Mean"], "After Median": row[f"S{i+1} Median"], "After Mode": row[f"S{i+1} Mode"],
            "After Std Dev": row[f"S{i+1} Std Dev"], "After Skewness": row[f"S{i+1} Skewness"], "After Kurtosis": row[f"S{i+1} Kurtosis"],
            "After Variance": row[f"S{i+1} Variance"], "After Max": row[f"S{i+1} Max"], "After Min": row[f"S{i+1} Min"], 
            "After Z-Score": row[f"S{i+1} Z-Score"], "After Mean STA/LTA": row[f"S{i+1} Mean STA/LTA"],
            "After Max STA/LTA": row[f"S{i+1} Max STA/LTA"], "After Mean MER": row[f"S{i+1} Mean MER"], "After Max MER": row[f"S{i+1} Max MER"],

            "earthquake": 0  # Label for noise data
        }
        noise_rows.append(new_row)

# Convert list to DataFrame
df_noise_fixed = pd.DataFrame(noise_rows)

# Add labels for earthquake data
df_earthquake["earthquake"] = 1

# Select only necessary columns from earthquake dataset
selected_columns = list(df_noise_fixed.columns)
df_earthquake_cleaned = df_earthquake[selected_columns]

# Combine datasets
df_combined = pd.concat([df_earthquake_cleaned, df_noise_fixed], ignore_index=True)

# Save the cleaned dataset
output_file = "/merged_dataset_with_sta_lta_and_mer.csv"
df_combined.to_csv(output_file, index=False)

print(f"Dataset merged successfully and saved as '{output_file}'.")
```
