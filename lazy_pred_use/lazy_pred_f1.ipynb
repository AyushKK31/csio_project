```python
# Install necessary libraries if not installed
# pip install pandas numpy scikit-learn lazypredict

import pandas as pd
from sklearn.model_selection import train_test_split
from lazypredict.Supervised import LazyClassifier

# Load both datasets
file_earthquake = "acceleration_stats_250_samples.csv"  # Update with actual path
file_noise = "noise_acceleration_stats_250_samples.csv"  # Update with actual path

df_earthquake = pd.read_csv(file_earthquake)
df_noise = pd.read_csv(file_noise)

# Add labels: 1 for earthquake, 0 for noise
df_earthquake["earthquake"] = 1
df_noise["earthquake"] = 0

# Combine datasets
df_combined = pd.concat([df_earthquake, df_noise], ignore_index=True)

# Drop non-feature columns
df_combined = df_combined.drop(columns=["File", "Trace ID", "P-Pick Time"], errors="ignore")

# Separate features and target
X = df_combined.drop(columns=["earthquake"])
y = df_combined["earthquake"]

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and run LazyClassifier
clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
models, predictions = clf.fit(X_train, X_test, y_train, y_test)

# Display results
print(models)
```
